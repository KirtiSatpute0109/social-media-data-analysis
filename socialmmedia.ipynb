```python
import seaborn as sns  #https://seaborn.pydata.org/generated/seaborn.set_style.html
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
#We use rc i.e a runtime configuration (rc) containing the default styles for every plot element you create
matplotlib.rcParams['font.size'] = 13
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
#drop the null values
df=df.dropna()
```

```python
# Count the number of people who use social media
social_media_usage_counts = df['6. Do you use social media?'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 6))
social_media_usage_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99'])
plt.title('Percentage of People Who Use Social Media')
plt.ylabel('')
plt.show()
```

```python
# Count the number of each gender
gender_counts = df['2. Gender'].value_counts()

# Plot the bar chart
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['#66b3ff', '#ff9999'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Convert the '8. What is the average time you spend on social media every day?' column to numerical values
time_mapping = {
    'Less than an Hour': 0.5,
    'Between 1 and 2 hours': 1.5,
    'Between 2 and 3 hours': 2.5,
    'Between 3 and 4 hours': 3.5,
    'Between 4 and 5 hours': 4.5,
    'More than 5 hours': 6
}
df['Average Time Spent (hours)'] = df['8. What is the average time you spend on social media every day?'].map(time_mapping)

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Average Time Spent (hours)', y='18. How often do you feel depressed or down?', data=df)
plt.title('Correlation between Time Spent on Social Media and Feelings of Depression')
plt.xlabel('Average Time Spent on Social Media (hours)')
plt.ylabel('Feelings of Depression (1-5)')
plt.show()
```

```python
# Count the responses for the question about feelings towards comparisons
comparison_feelings_counts = df['16. Following the previous question, how do you feel about these comparisons, generally speaking?'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 6))
comparison_feelings_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Feelings about Comparisons through Social Media')
plt.ylabel('')
plt.show()
```

```python

# Plot the histogram
plt.figure(figsize=(8, 6))
df['19. On a scale of 1 to 5, how frequently does your interest in daily activities fluctuate?'].plot(kind='hist', bins=5, color='#66b3ff', edgecolor='black')
plt.title('Frequency of Interest Fluctuation in Daily Activities')
plt.xlabel('Interest Fluctuation (1-5)')
plt.ylabel('Count')
plt.show()
```

```python
# Count the responses for the question about seeking validation
validation_counts = df['17. How often do you look to seek validation from features of social media?'].value_counts()

# Plot the bar chart
plt.figure(figsize=(8, 6))
validation_counts.plot(kind='bar', color='#66b3ff')
plt.title('How Often Users Seek Validation from Social Media')
plt.xlabel('Frequency (1-5)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Group by gender and calculate the mean of the '12. On a scale of 1 to 5, how easily distracted are you?' column
gender_distraction = df.groupby('2. Gender')['12. On a scale of 1 to 5, how easily distracted are you?'].mean()

# Plot the bar chart
plt.figure(figsize=(10, 6))
gender_distraction.plot(kind='bar', color=['#66b3ff', '#ff9999', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6', '#c4e17f'])
plt.title('Average Distraction Level by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Distraction Level (1-5)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by age and count the number of users for each age
age_social_media_usage = df.groupby('1. What is your age?')['6. Do you use social media?'].value_counts().unstack().fillna(0)

# Plot the bar chart
plt.figure(figsize=(16,32))
age_social_media_usage.plot(kind='bar', stacked=True, color=['#66b3ff', '#ff9999'])
plt.title('Social Media Usage by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Do you use social media?', loc='upper right')
plt.show()
```

```python

# Group by occupation status and calculate the mean of the 'Average Time Spent (hours)' column
occupation_time_spent = df.groupby('4. Occupation Status')['Average Time Spent (hours)'].mean()

# Plot the bar chart
plt.figure(figsize=(12, 6))
occupation_time_spent.plot(kind='bar', color='#66b3ff')
plt.title('Average Time Spent on Social Media by Occupation Status')
plt.xlabel('Occupation Status')
plt.ylabel('Average Time Spent (hours)')
plt.xticks(rotation=45)
plt.show()
```

```python

# Group by relationship status and calculate the mean of the 'Average Time Spent (hours)' column
relationship_time_spent = df.groupby('3. Relationship Status')['Average Time Spent (hours)'].mean()

# Plot the bar chart
plt.figure(figsize=(12, 6))
relationship_time_spent.plot(kind='bar', color='#66b3ff')
plt.title('Average Time Spent on Social Media by Relationship Status')
plt.xlabel('Relationship Status')
plt.ylabel('Average Time Spent (hours)')
plt.xticks(rotation=45)
plt.show()
```

```python

# Group by age and calculate the mean of the '18. How often do you feel depressed or down?' column
age_depression = df.groupby('1. What is your age?')['18. How often do you feel depressed or down?'].mean()

# Plot the line chart
plt.figure(figsize=(12, 6))
age_depression.plot(kind='line', marker='o', color='#66b3ff')
plt.title('Average Feelings of Depression by Age')
plt.xlabel('Age')
plt.ylabel('Average Feelings of Depression (1-5)')
plt.grid(True)
plt.show()
```

```python
# Count the responses for the question about comparing oneself to others
comparison_counts = df['15. On a scale of 1-5, how often do you compare yourself to other successful people through the use of social media?'].value_counts()

# Plot the bar chart
plt.figure(figsize=(8, 6))
comparison_counts.plot(kind='bar', color='#66b3ff')
plt.title('Frequency of Comparing Oneself to Others through Social Media')
plt.xlabel('Frequency (1-5)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python

# Group by gender and calculate the mean of the '19. On a scale of 1 to 5, how frequently does your interest in daily activities fluctuate?' column
gender_interest_fluctuation = df.groupby('2. Gender')['19. On a scale of 1 to 5, how frequently does your interest in daily activities fluctuate?'].mean()

# Plot the bar chart
plt.figure(figsize=(10, 6))
gender_interest_fluctuation.plot(kind='bar', color=['#66b3ff', '#ff9999', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6', '#c4e17f'])
plt.title('Average Interest Fluctuation in Daily Activities by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Interest Fluctuation (1-5)')
plt.xticks(rotation=45)
plt.show()
```
